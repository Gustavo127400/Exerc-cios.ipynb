# **II. Adicione uma nova coluna ao DataFrame chamada valor_total que seja o resultado da multiplicação da quantidade pelo preço unitário. Valor total vendido por produto:**

````python
 df_vendas['valor_total'] = df_vendas['quantidade'] * df_vendas['preco_unitario'] # Nova váriavel que calcula o valor total.
df_vendas
````

# III.  **Agrupe o DataFrame por produto e calcule a soma dos valores totais de cada produto. Total de produtos vendidos:**

````python
count_produtos = df_vendas.groupby('produto')['valor_total'].count() # count(): Conta o número de ocorrências.
mean_produtos = df_vendas.groupby('produto')['valor_total'].mean()# mean(): Calcula a média dos valores.
min_produtos = df_vendas.groupby('produto')['valor_total'].min()# min(): Encontra o valor mínimo.
max_produtos = df_vendas.groupby('produto')['valor_total'].max()# max(): Encontra o valor máximo.
std_produtos = df_vendas.groupby('produto')['valor_total'].std()# std(): Calcula o desvio padrão.
sum_produtos = df_vendas.groupby('produto')['valor_total'].sum() # sum(): Soma todos os valores.

sum_produtos
````

# **IV. Calcule o total de produtos vendidos somando todas as quantidades. Cliente que mais gastou:**

````python
sum_clientes = df_vendas.groupby('id_cliente')['valor_total'].sum()
print(sum_clientes)

print("Segundo nosso dataFrame(IV), o id do cliente que mais gastou foi: {}".format((sum_clientes.idxmax())))
````
